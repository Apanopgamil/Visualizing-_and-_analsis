```python
import pandas as pd
import numpy as nb 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('dailyActivity_merged.csv')
df.head()
```

```python
df.info()
```

```python
newdf =['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','Calories']
df=df[newdf]
df
```

```python
df['total_mintes']=df.VeryActiveMinutes+df.FairlyActiveMinutes+df.LightlyActiveMinutes
df.head()
```

```python
df.info()
```

```python
df.ActivityDate=pd.to_datetime(df.ActivityDate)
```

```python
import datetime as dt
df['dayofweek']=df.ActivityDate.dt.day_name()
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=((6,5)))
plt.hist(df.dayofweek,bins=7,color='red',width=0.6)
plt.xlabel('dayofweek')
plt.ylabel('frquency')
plt.title('frquency byday of week')
plt.show()
```
