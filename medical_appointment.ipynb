```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df =pd.read_csv('medical_appointment.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
x=df.query("Age==-1")
x
```

```python
df.drop(index=99832,inplace=True)
```

```python
df.rename(columns={'Hipertension':'Hypertension'},inplace=True)
df.rename(columns={'No-show':'No_show'},inplace=True)
```

```python
df.drop_duplicates(['PatientId','No_show'],inplace=True)
df.shape
```

```python
df.drop(['PatientId','AppointmentID','ScheduledDay','AppointmentDay'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.hist(figsize=(16,6));
```

```python
show=df.No_show=="No"
noshow=df.No_show=='Yes'
df[show].count(),df[noshow].count()
```

```python
df[show].mean(),df[noshow].mean()
```

```python
def attendance(df,col_name,attended,absent):
    plt.figure(figsize=[16,4])
    df[col_name][show].hist(alpha=.5,bins=10,label='show')
    df[col_name][noshow].hist(alpha=.5,bins=10,label='noshow')
    plt.legend();
    plt.title('comperison acc to age')
    plt.xlabel('Age')
    plt.ylabel('Patients Number');
attendance(df,"Age",show,noshow)
```

```python
plt.figure(figsize=[16,4])
df[show].groupby(['Hypertension','Diabetes']).mean()['Age'].plot(kind='bar',color='red',label='show')
df[noshow].groupby(['Hypertension','Diabetes']).mean()['Age'].plot(kind='bar',color='blue',label='noshow')
plt.legend();
plt.title('comperison acc to age  chronic diseaes')
plt.xlabel('chronic diseaes')
plt.ylabel('mean age')
```

```python
df[show].groupby(['Hypertension','Diabetes']).mean()['Age'],df[noshow].groupby(['Hypertension','Diabetes']).mean()['Age']
```

```python
def attendance(df,col_name,attended,absent):
    plt.figure(figsize=[12,4])
    df[col_name][show].value_counts(normalize=True).plot(kind='pie',label='show')
    plt.legend();
    plt.title('comperison between Gender')
    plt.xlabel('Gender')
    plt.ylabel('Patients Number');
attendance(df,"Gender",show,noshow)
```

```python
def attendance(df,col_name,attended,absent):
    plt.figure(figsize=[16,4])
    df[col_name][show].hist(alpha=.5,color='red',bins=15,label='show')
    df[col_name][noshow].hist(alpha=.5,color='blue',bins=15,label='noshow')
    plt.legend();
    plt.title('comperison acc receiving massage')
    plt.xlabel('SMS')
    plt.ylabel('Patients Number');
attendance(df,"SMS_received",show,noshow)
```

```python
plt.figure(figsize=[16,4])
df.Neighbourhood[show].value_counts().plot(kind='bar',color='red',label='show')
df.Neighbourhood[noshow].value_counts().plot(kind='bar',color='blue',label='noshow')
plt.legend();
plt.title('comperison acc to neightbourhood')
plt.xlabel('neightbourhood')
plt.ylabel('patients numper')
```

```python
plt.figure(figsize=[18,3])
df[show].groupby('Neighbourhood').SMS_received.mean().plot(kind='bar',color='red',label='show')
df[noshow].groupby('Neighbourhood').SMS_received.mean().plot(kind='bar',color='blue',label='noshow')
plt.legend();
plt.title('comperison acc to neightbourhood, sms resiving')
plt.xlabel('neightbourhood')
plt.ylabel('patients numper')
```

```python
plt.figure(figsize=[18,3])
df[show].groupby('Neighbourhood').Age.mean().plot(kind='bar',color='red',label='show')
df[noshow].groupby('Neighbourhood').Age.mean().plot(kind='bar',color='blue',label='noshow')
plt.legend();
plt.title('comperison acc to neightbourhood, Age')
plt.xlabel('neightbourhood')
plt.ylabel('patients numper')
```
